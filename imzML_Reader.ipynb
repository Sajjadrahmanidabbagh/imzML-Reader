```python
# imzML Reader and Image Visualization

import numpy as np
import matplotlib.pyplot as plt
from pyimzml.ImzMLParser import ImzMLParser

# 1. Define the file path
file_path = r"...centroid.imzML"

# 2. Parse the imzML file
print(f"Loading {file_path}...")
p = ImzMLParser(file_path)

# 3. Determine the image dimensions from the coordinates
# pyimzML coordinates are typically (x, y, z)
max_x = max(coords[0] for coords in p.coordinates)
max_y = max(coords[1] for coords in p.coordinates)

# Initialize a 2D numpy array with zeros
# Note: shape is (rows, columns) which corresponds to (y, x)
tic_image = np.zeros((max_y, max_x))

# 4. Populate the image array with Total Ion Current (TIC)
print("Extracting spectra and building the image...")
for idx, (x, y, z) in enumerate(p.coordinates):
    # Retrieve m/z values and their corresponding intensities for the current pixel
    mzs, intensities = p.getspectrum(idx)
    
    # Calculate TIC by summing all intensities at this pixel
    # We subtract 1 from x and y because imzML coordinates usually start at 1, 
    # whereas numpy arrays are 0-indexed.
    tic_image[y - 1, x - 1] = np.sum(intensities)

tic_image_corrected = np.rot90(tic_image, k=1)

# 5. Visualize the resulting image
plt.figure(figsize=(10, 8))

# Make sure to plot the CORRECTED image here
plt.imshow(tic_image_corrected, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Total Ion Current (TIC)')
plt.title('imzML Alignment Check (Corrected Orientation)')

# Note: Because you rotated the image, the X and Y labels technically swap relative to the raw data
plt.xlabel('Original Y Coordinate axis')
plt.ylabel('Original X Coordinate axis')

plt.grid(False) 
plt.show()
```
